<font size="14">Checking for any missing parameters</font>

```python
network.buses
```

```python
# Check if there are any missing carriers and set a default
missing_buses = network.buses['carrier'].isna()
missing_lines = network.lines['carrier'].isna()

# Set default carrier for missing ones
network.buses.loc[missing_buses, 'carrier'] = 'AC'

network.lines.loc[missing_lines, 'carrier'] = 'AC'

# Ensure 'AC' carrier is defined in network
if 'AC' not in network.carriers.index:
    network.add("Carrier", "AC")
```

```python
# Define the number of snapshots and group size
num_snapshots = len(network.snapshots)  # Get total number of snapshots
group_size = 24  # Set group size, e.g., 24 snapshots (if each snapshot represents one hour)

# Optimize snapshots in groups
for i in range(int(num_snapshots / group_size)):
    snapshots_subset = network.snapshots[group_size * i : group_size * i + group_size]
    network.optimize(
        snapshots_subset,
        solver_name="highs",
    )
```

```python
network.objective
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString
import contextily as ctx

# ----------------------------
# 1. Build buses GeoDataFrame
# ----------------------------
if {"x", "y"}.issubset(network.buses.columns):
    buses_gdf = gpd.GeoDataFrame(
        network.buses,
        geometry=gpd.points_from_xy(network.buses.x, network.buses.y),
        crs="EPSG:4326"   # WGS84 lat/lon
    )
elif {"lon", "lat"}.issubset(network.buses.columns):
    buses_gdf = gpd.GeoDataFrame(
        network.buses,
        geometry=gpd.points_from_xy(network.buses.lon, network.buses.lat),
        crs="EPSG:4326"
    )
else:
    raise ValueError("Bus coordinates not found! Expected (x,y) or (lon,lat).")

# ----------------------------
# 2. Build line geometries
# ----------------------------
line_geometries = []
for _, line in network.lines.iterrows():
    if line.bus0 in buses_gdf.index and line.bus1 in buses_gdf.index:
        p0 = buses_gdf.loc[line.bus0].geometry
        p1 = buses_gdf.loc[line.bus1].geometry
        line_geometries.append(LineString([p0, p1]))
    else:
        line_geometries.append(None)

lines_gdf = gpd.GeoDataFrame(
    network.lines,
    geometry=line_geometries,
    crs="EPSG:4326"
)

# ----------------------------
# 3. Reproject to Web Mercator (needed for basemap)
# ----------------------------
buses_gdf_3857 = buses_gdf.to_crs(epsg=3857)
lines_gdf_3857 = lines_gdf.to_crs(epsg=3857)

# ----------------------------
# 4. Plot with basemap
# ----------------------------
fig, ax = plt.subplots(figsize=(16, 12))

# Plot lines
lines_gdf_3857.plot(ax=ax, color="blue", linewidth=1, alpha=0.6, label="Lines")

# Plot buses
buses_gdf_3857.plot(ax=ax, color="red", markersize=30, label="Buses")

# Add OSM basemap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

ax.set_title("Network Topology with Basemap", fontsize=14)
ax.legend()
plt.show()
```

```python

# Extract active and reactive power at the from-bus end
P = network.lines_t["p0"]   # DataFrame (snapshots  lines)
Q = network.lines_t["q0"]

# Compute apparent power magnitude
S0 = np.sqrt(P**2 + Q**2)


# Print first few values
print(S0.head())
```

```python
TotalPower_line=network.lines_t.q0.abs()
# Define the output CSV file path
csv_filename_all_snapshots =os.path.join(data_folder,'output1', f'TotalPower_line.csv')

# Save the collected data to a single CSV file
TotalPower_line.to_csv(csv_filename_all_snapshots, index=True)

# Print confirmation
print(f"Line loadings saved to {csv_filename_all_snapshots}")
```

```python
line_loading_final = S0.div(network.lines.s_nom, axis=1) * 100

# Save
csv_filename_all_snapshots = os.path.join(data_folder,'output1','line_loading_final.csv')
line_loading_final.to_csv(csv_filename_all_snapshots, index=True)

print(f"Line loadings saved to {csv_filename_all_snapshots}")
print(line_loading_final.head())
```

```python
line_loading_all = S0.abs().div(network.lines.s_nom, axis=1) * 100
# Example: plot first 5 lines across all snapshots
plt.figure(figsize=(14, 6))
line_loading_all.iloc[:, :5].plot(ax=plt.gca())

plt.xlabel("Snapshot")
plt.ylabel("Loading (%)")
plt.title("Line Loading Over Time")
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Calculate line loading (%) for chosen snapshot
hour_to_analyze = network.snapshots[20]  # Select the 21st snapshot

line_loading_final_hourly = (
    S0.abs().loc[hour_to_analyze].div(network.lines.s_nom) * 100
)

# Find overloaded lines (>100%)
overloaded_lines = line_loading_final_hourly[line_loading_final_hourly > 100]

print("Overloaded lines at", hour_to_analyze)
print(overloaded_lines)
```

<font size="14">Load Distributions on Based Map</font>

```python
import matplotlib.pyplot as plt
import contextily as ctx
from matplotlib.collections import LineCollection
import numpy as np

# ----------------------------
# Choose snapshot
# ----------------------------
hour_to_analyze = network.snapshots[20]   # example: 21st hour
line_loading_snapshot = line_loading_final.loc[hour_to_analyze]

# ----------------------------
# Reproject for basemap
# ----------------------------
lines_gdf_3857 = lines_gdf.to_crs(epsg=3857)
buses_gdf_3857 = buses_gdf.to_crs(epsg=3857)

# ----------------------------
# Line segments + coloring
# ----------------------------
lines_coords = [np.array(geom.coords) for geom in lines_gdf_3857.geometry]

line_segments = LineCollection(
    lines_coords,
    linewidths=2,
    cmap="viridis",
    norm=plt.Normalize(vmin=0, vmax=max(100, line_loading_snapshot.max())),
    array=line_loading_snapshot.values
)

# ----------------------------
# Plot
# ----------------------------
fig, ax = plt.subplots(figsize=(12, 12))

# Add lines
ax.add_collection(line_segments)

# Add buses
buses_gdf_3857.plot(ax=ax, color="red", markersize=30, label="Buses")

# Add OSM basemap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

# Add colorbar
cbar = plt.colorbar(line_segments, ax=ax, orientation="vertical", fraction=0.03, pad=0.02)
cbar.set_label("Line Loading (%)")

# ----------------------------
# Annotate line loadings
# ----------------------------
for (geom, loading) in zip(lines_gdf_3857.geometry, line_loading_snapshot.values):
    if geom is not None:
        # midpoint of line
        x, y = geom.interpolate(0.5, normalized=True).coords[0]
        ax.text(
            x, y,
            f"{loading:.1f}%",     # one decimal percentage
            fontsize=7,
            color="black",
            ha="center",
            va="center",
            bbox=dict(facecolor="white", edgecolor="none", alpha=0.6, pad=0.5)  # white box for readability
        )

ax.set_title(f"Network Line Loadings at {hour_to_analyze}", fontsize=14)
ax.legend()
plt.show()
```
