<font size="14">Network on Map</font>

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString
import contextily as ctx

# ----------------------------
# 1. Build buses GeoDataFrame
# ----------------------------
if {"x", "y"}.issubset(network.buses.columns):
    buses_gdf = gpd.GeoDataFrame(
        network.buses,
        geometry=gpd.points_from_xy(network.buses.x, network.buses.y),
        crs="EPSG:4326"   # WGS84 lat/lon
    )
elif {"lon", "lat"}.issubset(network.buses.columns):
    buses_gdf = gpd.GeoDataFrame(
        network.buses,
        geometry=gpd.points_from_xy(network.buses.lon, network.buses.lat),
        crs="EPSG:4326"
    )
else:
    raise ValueError("Bus coordinates not found! Expected (x,y) or (lon,lat).")

# ----------------------------
# 2. Build line geometries
# ----------------------------
line_geometries = []
for _, line in network.lines.iterrows():
    if line.bus0 in buses_gdf.index and line.bus1 in buses_gdf.index:
        p0 = buses_gdf.loc[line.bus0].geometry
        p1 = buses_gdf.loc[line.bus1].geometry
        line_geometries.append(LineString([p0, p1]))
    else:
        line_geometries.append(None)

lines_gdf = gpd.GeoDataFrame(
    network.lines,
    geometry=line_geometries,
    crs="EPSG:4326"
)

# ----------------------------
# 3. Reproject to Web Mercator (needed for basemap)
# ----------------------------
buses_gdf_3857 = buses_gdf.to_crs(epsg=3857)
lines_gdf_3857 = lines_gdf.to_crs(epsg=3857)

# ----------------------------
# 4. Plot with basemap
# ----------------------------
fig, ax = plt.subplots(figsize=(16, 12))

# Plot lines
lines_gdf_3857.plot(ax=ax, color="blue", linewidth=1, alpha=0.6, label="Lines")

# Plot buses
buses_gdf_3857.plot(ax=ax, color="red", markersize=30, label="Buses")

# Add OSM basemap
ctx.add_basemap(ax, source=ctx.providers.OpenStreetMap.Mapnik)

ax.set_title("Network Topology with Basemap", fontsize=14)
ax.legend()
plt.show()
```

<font size="12">Transformer Thermal Loading Based on Current for a particulat DT</font>

```python
def transformer_loading_I(network, bus_voltages, transformer):
    """
    Thermal loading of transformer based on current magnitude.
    Uses LV side voltage (bus1).
    """
    tr = network.transformers.loc[transformer]
    bus_lv = tr.bus1         # LV side
    S_nom = tr.s_nom         # MVA rating
    V_lv_nom = network.buses.loc[bus_lv, "v_nom"]  # kV nominal
    
    # Rated current (A)
    I_rated = (S_nom * 1e3) / (np.sqrt(3) * V_lv_nom)
    
    # Complex power on LV side (convert MW/MVAr → kVA)
    P = network.transformers_t.p1[transformer]  # MW
    Q = network.transformers_t.q1[transformer]  # MVAr
    S_lv = (P + 1j*Q) * 1e3                     # kVA
    
    # Actual LV bus voltage (kV)
    V_lv = bus_voltages[bus_lv] * V_lv_nom
    
    # Current magnitude (A)
    I_lv = np.abs(S_lv) / (np.sqrt(3) * V_lv)
   #  I_lv = (S_lv) / (np.sqrt(3) * V_lv)

    # Loading %
    loading_pct = (I_lv / I_rated) * 100
    return loading_pct


# 🔹 Define transformer here (outside function)
transformer = "PB_DT_25"

# 🔹 Compute loading
loading_I = transformer_loading_I(network, bus_voltages, transformer)

# 🔹 Plot for one day
day = "2023-05-12"
plt.figure(figsize=(14,6))
plt.plot(loading_I.loc[day].index, loading_I.loc[day].values, marker="s", label="Loading % (I-based)")
plt.xlabel("Hour")
plt.ylabel("Loading (%)")
plt.title(f"Transformer {transformer} Loading on {day}")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd

def transformer_loading_S(network, transformer):
    """
    Thermal loading of transformer based on apparent power S (MVA).
    """
    # Transformer rating
    S_nom = network.transformers.loc[transformer, "s_nom"]  # MVA
    
    # Complex power flows (use LV side, bus1)
    P = network.transformers_t.p0[transformer]  # MW
    Q = network.transformers_t.q0[transformer]  # MVAr
    S = np.sqrt(P**2 + Q**2)                   # MVA magnitude
    
    # Loading %
    loading_pct = (S / S_nom) * 100
    return loading_pct
transformer = "PB_DT_25"   # pick your transformer

# Loading based on apparent power
loading_S = transformer_loading_S(network, transformer)

# Plot for one day
day = "2023-05-12"

plt.figure(figsize=(14,6))
plt.plot(loading_S.loc[day].index, loading_S.loc[day].values, marker="o", label="Loading % (S/s_nom)")
#plt.plot(loading_I.loc[day].index, loading_I.loc[day].values, marker="s", label="Loading % (I-based)")
plt.xlabel("Hour")
plt.ylabel("Loading (%)")
plt.title(f"Transformer {transformer} Loading on {day}")
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```
